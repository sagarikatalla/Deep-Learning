```
#select pop. estimate as response,Y rest are all X
dataset = df.values
X = dataset[:, 5:475]
y = dataset[:,-1]
y
```

```
index = np.arange(len(dataset))
splits = np.array_split(index, 10)
splits
index
```

```

from keras import models
from keras import layers
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D

def build_model():
    # Because we will need to instantiate
    # the same model multiple times,
    # we use a function to construct it.
    model = models.Sequential()
    model.add(layers.Dense(940, input_dim=470, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(470, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(235, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(117, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(58, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense( 1, activation='linear'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse','mae'])
    return model
  
  
   
```

#Model building

```
####### Model building ###############
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error



rf_model = RandomForestRegressor() 

 # Set up the whole range of indices in the dataset, and split it into ten
index = np.arange(len(dataset))
splits = np.array_split(index, 10)
splits

# k-fold cross-validation
fold = 1
all_scores = []
all_mae_train_histories = []
all_mae_histories = []
all_mse_train_histories = []
all_mse_histories = []

for i, split in enumerate(splits):
  print("Fold:", fold); fold += 1
  
  # Generate test and train indices, similar to sklearn's method
  test_index = split
  train_index = np.concatenate(splits[:i] + splits[i+1:], axis=0)
  
  # Split into X and y
  X_train, X_test = X[train_index], X[test_index]
  y_train, y_test = y[train_index], y[test_index]
  
  # Build the Keras model (already compiled)
  model1 = build_model()
    
  ##validation metrics
  history = model1.fit(X_train, y_train,validation_data=(X_test, y_test),epochs=200, batch_size=100, verbose=0)
  mae_history = history.history['val_mean_absolute_error']
  mse_history = history.history['val_mean_squared_error']
  
  all_mae_histories.append(mae_history)
  all_mse_histories.append(mse_history)

  ##training data error metrics
  history_train = model1.fit(X_train, y_train,validation_data=(X_train, y_train),epochs=200, batch_size=100, verbose=0)
  mae_train_history = history_train.history['val_mean_absolute_error']
  mse_train_history = history_train.history['val_mean_squared_error']
  
  all_mae_train_histories.append(mae_train_history)
  all_mse_train_histories.append(mse_train_history)
  
  
  # Evaluate neural net model on the train data
  nn_model_train  = [item for sublist in model1.predict(X_train) for item in sublist]
  df.loc[train_index, ('Neural Network'+' '+'Train')] = np.array(nn_model_train)
  
  # Evaluate neural net model on the validation data
  nn_model_test  = [item for sublist in model1.predict(X_test) for item in sublist]
  df.loc[test_index, ('Neural Network'+' '+'Test')] = np.array(nn_model_test)

  
  # Train each model
  rf_model.fit(X_train, y_train)
  #evaluate random forest model on train and test 
  df.loc[train_index, (str(rf_model)+'Train')] = rf_model.predict(X_train)
  df.loc[test_index, (str(rf_model)+'Test')] = rf_model.predict(X_test)
    
  ##bind fold column
  df.loc[train_index,'fold'] = fold
  df.loc[test_index,'fold'] = fold
```

#MSE plots for each fold

```
for i in range(10):
  import matplotlib.pyplot as plt
  test_mse = all_mse_histories[i]
  train_mse= all_mse_train_histories[i]
  plt.plot(range(1, len(train_mse) + 1), train_mse, color="#1033D2",  label="Training score")
  plt.plot(range(1, len(test_mse) + 1), test_mse, color="#E82E1A",  label="Test score")

 
  #creating plot
  plt.title("Train Vs. Test MSE for fold " + str(i+1))
  plt.legend(loc="best")

  plt.xlabel('Epochs')
  plt.ylabel('MSE')
  plt.tight_layout()

  plt.show()
```

#Summary table of each model (training and validation results)

```
def absolute_percentage_error(y_true, y_pred):
  return np.abs((y_true - y_pred)/y_true) * 100

y_true = df["medianPopulation"].values
error_arr = []
models = ['Neural Network'+' ',rf_model]
# Collect error metrics for each model
for model in models:
  train_mape = np.mean(absolute_percentage_error(y_true, df[str(model)+'Train'].values))
  train_ape_sd = np.std(absolute_percentage_error(y_true, df[str(model)+'Train'].values))
  train_mae = mean_absolute_error(y_true, df[str(model)+'Train'].values)
  train_ae_sd = np.std(np.abs(y_true - df[str(model)+'Train'].values))
  
  test_mape = np.mean(absolute_percentage_error(y_true, df[str(model)+'Test'].values))
  test_ape_sd = np.std(absolute_percentage_error(y_true, df[str(model)+'Test'].values))
  test_mae = mean_absolute_error(y_true, df[str(model)+'Test'].values)
  test_ae_sd = np.std(np.abs(y_true - df[str(model)+'Test'].values))
  
  error_arr.append([train_mape, train_ape_sd, train_mae, train_ae_sd,
                    test_mape, test_ape_sd, test_mae, test_ae_sd])
  
# Create table
error_df = pd.DataFrame(error_arr, columns = ['Train MAPE', '(sd)', 'Test MAPE', '(sd)', 'Train MAE', '(sd)', 'Test MAE', '(sd)'])
error_df.rename(index={0:'Neural Network', 1:'Random Forest'}, inplace=True)
display(error_df)
```

# Boxplots

```
import matplotlib.pyplot as plt

# Calculate the mean test prediction between all three models
df['mean_test_pred'] = np.mean([df[str(model)+'Test'].values for model in models])

states = ['California', 'Texas', 'Florida']

# Generate AE and APE vectors for each state
state_ae_arr = []
state_ape_arr = []
for state in states:
  df_state = df[df['STNAME'] == state]
  y_true = df_state['medianPopulation'].values
  y_pred = df_state['mean_test_pred'].values
  
  state_ae_arr.append(np.abs(y_true - y_pred))
  state_ape_arr.append(absolute_percentage_error(y_true, y_pred))

# Box plot for AE (no outliers)
fig, ax = plt.subplots()
ax.boxplot(state_ae_arr, showfliers=False)
ax.set_xlabel('State')
ax.set_ylabel('AE')
ax.set_xticklabels(states)

plt.show()
```

```
# Box plot for APE (no outliers)
fig, ax = plt.subplots()
ax.boxplot(state_ape_arr, showfliers=False)
ax.set_xlabel('State')
ax.set_ylabel('APE')
ax.set_xticklabels(states)

plt.show()
```
